This notebook analyzes job roles, locations, and salary patterns in the Indian job market.

```
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('../data/dataset.csv')

# Display first 5 rows
df.head()
```

```
# Job count by location
location_counts = df['Location'].value_counts()
location_counts
```

```
# Plot job distribution by location
location_counts.plot(kind='bar', title='Job Distribution by Location')
plt.xlabel('Location')
plt.ylabel('Number of Jobs')
plt.show()
```
